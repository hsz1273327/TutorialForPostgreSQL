# OLTP场景


OLTP场景主要是为了应对高并发要求.pg天生为OLTP场景设计,在单机情景下,PG的性能指标可以随TPS维持双曲线甚至对数曲线上升,到顶峰之后不再下降.同时支持分区操作,可以进一步优化插入和简单查询性能;也可以使用主备模式做按业务区分的读写分离,使用主备模式做高可用.

除此之外pg可以直接查询数据库的自身状态,以便于在需要时进行优化.

OLTP场景下一般数据库也都是就近部署,这是因为通常OLTP任务更贴近业务,而业务在很大程度上会受落地地区的政策影响.在就近部署的条件下通常单节点就已经足够.下面是单机PostgreSQL的理论极限容量:

项目|限制
---|---
最大单个数据库大小|不限	
最大数据单表大小|32 TB
单条记录最大|1.6 TB
单字段最大允许|1 GB
单表允许最大记录数|不限
单表最大字段数|250 - 1600 (取决于字段类型)
单表最大索引数|不限

正常情况下只要业务数据设计合理很少会有业务数据能达到pg的容量极限.

当然了更多的情况是业务还没能走出国门就黄了,从避免过早优化的角度看OLTP场景主要也应该以单节点部署为主.但事情总会有例外,我们总归会需要给自己留个后路,万一真的发展到了极大的规模,单节点就是无法满足怎么办?因此集群部署依然是需要考虑的问题.
